```python
X = df.drop('target',axis=1)
Y = df['target']

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2)
```

```python
k_range = list(range(1,4))
acc_list = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    acc = accuracy_score(y_test,y_pred)*100
acc_list.append(acc)
print('Accuracy of KNN Model is:',acc_list)

```

```python
knn.compile(optimizer=keras.optimizers.Adam(), loss = 'binary_crossentropy', metrics=['accuracy'])

knn.summary()
```

```python
knn.fit(x_train,y_train,epochs=5)
```

```python
loss, accuracy = knn.evaluate(x_test,y_test)
print('loss=',loss*100)
print('Accuracy=',accuracy*100)
```
